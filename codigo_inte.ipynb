##Trayectoria del vector relativo en el sitena perifocal

```python
p = 1.0
e = 0.7
theta = np.linspace(0, 2*np.pi, 100)
rs= p / (1 + e* np.cos(theta))

xfs = rs * np.cos(theta)
yfs = rs * np.sin(theta)
# pondremos el origen en el foco

plt.plot(xfs, yfs)
plt.plot(0, 0, 'yo', label='Sun')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Orbit of a planet around the sun')
plt.grid()
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
#Dibuja la órbita de un planeta alrededor del sol
import ipywidgets as widgets
from ipywidgets import interact

def dibuja_orbita(e):
    p = 1.0
    q=p
    Q=p
    theta = np.linspace(0, 2*np.pi, 100)
    rs = p / (1 + e * np.cos(theta))

    xfs = rs * np.cos(theta)
    yfs = rs * np.sin(theta)
    plt.figure(figsize=(6, 6))
    plt.plot(xfs, yfs)

    plt.plot(0, 0, 'yo', label='Sun')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Orbit of a planet around the sun')
    plt.xlim(-2*Q, 2*Q)
    plt.ylim(-2*Q, 2*Q)
    plt.grid()
    
    plt.tight_layout()
    plt.show()
```

```python
interact(dibuja_orbita, e=(-1.5, 1.5, 0.1))
```

```python
m1 = 1
m2 = 1
M = m1 + m2

rsvec = np.array([xfs, yfs]).T

r1svec = m2 / M * rsvec
r2svec =- m1 / M * rsvec
```

```python
plt.plot(r1svec[: , 0], r1svec[: , 1], label='body 1')
plt.plot(r2svec[: , 0], r2svec[: , 1], label='body 2')
plt.plot(0, 0, 'yo', label='Sun')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Orbit of two bodies around the sun')
plt.grid()
plt.axis('equal')
plt.legend()
plt.show()
```
