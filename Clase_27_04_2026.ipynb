```python
import pymcel as pc
import numpy as np
import matplotlib.pyplot as plt
```

```python
deg = np.pi/180
```

```python
e = 0.1911663355386932 	
a= 0.9223803173917017 *pc.constantes.au
q = 0.7460522521429133	*pc.constantes.au
i = 3.340958441017069	*deg
node = 203.8996515621043*deg	
peri = 126.6728325163065*deg
M = 312.8054663584516	
tp = 2461042.918242006079 * 24 * 3600  #segundos
mu = pc.constantes.mu_sun  #todo el sistema solar
```

```python
#variables derivadas: necesitamos h
h = np.sqrt(mu*a*(1-e**2))
b = a*np.sqrt(1-e**2)
```

```python
from astropy.time import Time
```

```python
t = Time("2029-04-13 18:52:00", scale = 'tdb').jd
t
```

```python
#escribir la ecuacion de kepler en la forma f(E) = 0, con E el anomalia excéntrica
def ecuacion_kepler(E):
    funcion = E - e*np.sin(E) - h/(a*b)*(t-tp)
    return funcion
```

```python
#resolver la ecuacion de kepler con el metodo 
from scipy.optimize import newton
```

```python
#pasar de dias a segundos
t = t*24*3600
```
