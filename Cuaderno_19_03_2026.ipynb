```python
m_zul_Aa=1
m_zul_Ab= 0.01


r_zul_A= np.array([0.1,0,0])
v_zul_A= np.array([0,1,0])  
```

```python
m_zul_Aab = m_zul_Aa + m_zul_Ab
m_zul_B = 1.0



r_zul_AB= np.array([1,0,0])
v_zul_AB= np.array([0,1,0])
```

```python
# Armar el sistema jerarquico, primero hay que definir el centro de masa del sistema mas grande, o mayor (coordenadas de Jacobi)
RCM_zul_AB = np.array([0,0,0])
VCM_zul_AB = np.array([0,0,0])
```

Descendemos por el sistema del arbol jerarquico

```python

#Este es el equivalente a la particula 1
RCM_zul_A = RCM_zul_AB + m_zul_B/(m_zul_Aab + m_zul_B)*r_zul_AB
VCM_zul_A = VCM_zul_AB + m_zul_B/(m_zul_Aab + m_zul_B)*v_zul_AB

# este es el equyivalente a la particula 2
r_zul_B = RCM_zul_AB - m_zul_Aab/(m_zul_Aab + m_zul_B)*r_zul_AB
v_zul_B = VCM_zul_AB - m_zul_Aab/(m_zul_Aab + m_zul_B)*v_zul_AB

```

```python
#particula 1
r_zul_Aa = RCM_zul_A + m_zul_Ab/m_zul_Aab * r_zul_A
v_zul_Aa = VCM_zul_A + m_zul_Ab/m_zul_Aab * v_zul_A

#particula 2
r_zul_Ab = RCM_zul_A - m_zul_Aa/m_zul_Aab * r_zul_A
v_zul_Ab = VCM_zul_A - m_zul_Aa/m_zul_Aab * v_zul_A
```
